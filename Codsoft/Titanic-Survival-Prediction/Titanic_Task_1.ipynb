```python
## drop unnecessary columns
titanic_dataset = titanic_dataset.drop(columns=['Name', 'Ticket'], axis=1)
titanic_dataset.head()
```

Encoding Label

```python
titanic_dataset["Sex"].value_counts()
```

```python
titanic_dataset['Embarked'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Sex', 'Embarked']
le = LabelEncoder()

for col in cols:
    titanic_dataset[col] = le.fit_transform(titanic_dataset[col])
titanic_dataset.head()
```

Train Test Split

```python
X = titanic_dataset.drop(columns = ['PassengerId','Survived'],axis=1)
Y = titanic_dataset['Survived']
print(X)
```

```python
print(Y)
```

```python
##Splitting the data into training data & Test data.
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier(random_state=1)
```

```python
titanic_dataset.info()
```

```python
titanic_dataset.astype({'Age':'int','Fare':'int'}).dtypes
```

```python
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy_score_of_training_data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy_score_of_test data : ', test_data_accuracy)
```
